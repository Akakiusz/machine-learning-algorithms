## Who Survives the Titanic? A Decision Tree Analysis

### Project Overview

This project explores the factors that influenced passenger survival on the Titanic using a **Decision Tree** classification model.

###  Objectives

* Predict who survived or not
* Identify the most important factors which increesed survival
* Build a model using Decision Trees



```python
# First section with import libraries

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
```

```python
# Loading the data
data = pd.read_csv("../CSV data/titanic.csv")
print("data loaded successfully")
```

```python
# Line give us 5 raws over dataset
data.head(10) 
```

```python
# Features (x) and target (y) we can check how factors trigger machine failure (0/1)
x = data[[ "Age", "SibSp", "Parch", "Fare", "Pclass"]]
y = data["Survived"]
```

```python
# Run decision tree classifier 
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf = clf.fit(x, y)
```

```python
# Visualisation gini
plt.figure(figsize=(15,10))
tree.plot_tree(clf, feature_names=["Age", "SibSp", "Parch", "Fare", "Pclass"],
    class_names=["Survived 0" , "Survived 1"], filled=True, fontsize=10)
plt.show()
```

### Outcomes:

- People who paid low fare (<= 15.173) → mostly died
- Lower class (3rd class) → high death rate
- People with some family onboard (SibSp) sometimes had better survival
- But too many family members → worse outcomes
- Younger passengers sometimes had slightly better chances
- But your tree shows it is not the strongest factor (compared to Fare/Pclass)
